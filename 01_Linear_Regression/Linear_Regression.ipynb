```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices.csv")
df
```

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(df.Area,df.Price,color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
p=reg.predict(pd.DataFrame({'Area':[3300]}))
```

```python
p
```

```python
m=reg.coef_
m
```

```python
b=reg.intercept_
b
```

```python
y=m*3300+b
```

```python
y
```
